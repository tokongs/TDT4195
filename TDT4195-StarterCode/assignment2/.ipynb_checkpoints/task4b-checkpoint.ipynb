```python
import matplotlib.pyplot as plt
import numpy as np
import skimage
import utils
```

```python
def magnitude(fft_im):
    real = fft_im.real
    imag = fft_im.imag
    return np.sqrt(real**2 + imag**2)

def convolve_im(im: np.array,
                kernel: np.array,
                verbose=True):
    """ Convolves the image (im) with the spatial kernel (kernel),
        and returns the resulting image.

        "verbose" can be used for turning on/off visualization
        convolution
        
        Note: kernel can be of different shape than im.

    Args:
        im: np.array of shape [H, W]
        kernel: np.array of shape [K, K] 
        verbose: bool
    Returns:
        im: np.array of shape [H, W]
    """
    ### START YOUR CODE HERE ### (You can change anything inside this block)
    pad_y = (len(im) - len(kernel)) // 2
    pad_x = (len(im[0]) - len(kernel[0])) // 2 
    kernel = np.pad(kernel, (( pad_y+1, pad_y), (pad_x+1, pad_x)), mode='constant', constant_values=0)
    
    fft_kernel = np.fft.fft2(kernel)
    fft_kernel_viz = np.log(np.abs(fft_kernel)+1)
    
    fft_im = np.fft.fft2(im)
    fft_im_filtered = fft_im*fft_kernel
    
    
    fft_viz = np.fft.fftshift(fft_im)
    fft_viz = np.log(magnitude(fft_viz)+1)
    
    fft_filtered_viz = np.fft.fftshift(fft_im_filtered)
    fft_filtered_viz = np.log(magnitude(fft_filtered_viz)+1)
    conv_result = np.fft.ifft2(fft_im_filtered).real
    if verbose:
        # Use plt.subplot to place two or more images beside eachother
        plt.figure(figsize=(20, 4))
        # plt.subplot(num_rows, num_cols, position (1-indexed))
        plt.subplot(1, 5, 1)
        plt.imshow(im, cmap="gray")
        plt.subplot(1, 5, 2) 
        # Visualize FFT
        plt.imshow(fft_viz, cmap="gray")
        plt.subplot(1, 5, 3)
        # Visualize FFT kernel
        plt.imshow(fft_kernel_viz, cmap="gray")
        plt.subplot(1, 5, 4)
        # Visualize filtered FFT image
        plt.imshow(fft_filtered_viz, cmap="gray")
    
        plt.subplot(1, 5, 5)
        # Visualize filtered spatial image
        plt.imshow(conv_result, cmap="gray")

    ### END YOUR CODE HERE ###
    return conv_result
```

```python
if __name__ == "__main__":
    verbose = True  # change if you want

    # Changing this code should not be needed
    im = skimage.data.camera()
    im = utils.uint8_to_float(im)

    # DO NOT CHANGE
    gaussian_kernel = np.array([
        [1, 4, 6, 4, 1],
        [4, 16, 24, 16, 4],
        [6, 24, 36, 24, 6],
        [4, 16, 24, 16, 4],
        [1, 4, 6, 4, 1],
    ]) / 256
    image_gaussian = convolve_im(im, gaussian_kernel, verbose)

    # DO NOT CHANGE
    sobel_horizontal = np.array([
        [-1, 0, 1],
        [-2, 0, 2],
        [-1, 0, 1]
    ])
    image_sobelx = convolve_im(im, sobel_horizontal, verbose)

    if verbose:
        plt.show()

    utils.save_im("camera_gaussian.png", image_gaussian)
    utils.save_im("camera_sobelx.png", image_sobelx)
```
